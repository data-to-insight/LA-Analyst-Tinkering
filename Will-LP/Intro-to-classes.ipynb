```python
# We define a class like a function using class, then the name of the class
class Pet():
    # We define an __init__ functions (like a function), that tells the class what to do when you start
    # You generally use it to define the data items. Do this by setting 
    def __init__(self, animal, noise, name):
        self.animal = animal
        self.noise = noise
        self.name = name
    
    def speak(self):
        print(self.noise)
        
cat = Pet('Cat', 'Meow', 'Percy')
```

```python
print(cat.name)
cat.speak()
```

```python
import piplite
await piplite.install('seaborn')

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
class DataReader():
    def __init__(self, filepath):
        self.filepath = filepath
        self.data = self._read_files(filepath)
        
        
    def _read_files(self, filepath):
        try:
            data = pd.read_excel(self.filepath)
        except:
            data = pd.read_csv(self.filepath)
        return data
```

```python
ethnicity_data = DataReader('data/spc_pupils_ethnicity_and_language_.csv')
print(ethnicity_data.data)
```

```python
class DataReader2():
    def __init__(self, filepath):
        self.filepath = filepath
        self.data = self._read_files(filepath)
        
        
    def _read_files(self, filepath):
        try:
            data = pd.read_excel(self.filepath)
        except:
            data = pd.read_csv(self.filepath)
        return data
    
    def plotter(self, la, time_period):
        plt.clf()
        df = self.data
        df = df[df['phase_type_grouping'] == 'Total']
        df = df[df['ethnicity'] != 'Total']
        df = df[df['time_period'] == time_period]
        df = df[df['la_name'] == la]
        sns.barplot(df, x='ethnicity', y='headcount')
        plt.xticks(rotation=90)
        plt.subplots_adjust(bottom=0.4)
        plt.show()
        #print(df)
```

```python
to_plot = DataReader2('data/spc_pupils_ethnicity_and_language_.csv')
```

```python
to_plot.plotter('Luton', 202122)
```
